```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
```

```python
#엑셀 파일 불러오기
file_path = ('/content/drive/MyDrive/등록외국인+현황(연령별_동별)_20240602150702.xlsx')

# 엑셀 파일을 읽어옵니다
df = pd.read_excel(file_path, header=None)

# 두 번째 행을 헤더로 사용합니다 (두 번째 행을 인덱스로 설정하고 나머지 데이터프레임을 재구성)
df.columns = df.iloc[0]  # 두 번째 행을 열 이름으로 설정


# "남자"와 "여자"를 포함하는 행을 삭제합니다
df = df[~df.apply(lambda row: row.astype(str).str.contains("남자|여자").any(), axis=1)]
df = df.iloc[1:,:]

# 결과를 확인합니다
print(df)
```

```python
df
```

```python
# '항목'과 '계'가 포함된 열의 인덱스 확인
drop_columns = [col_idx for col_idx, col in enumerate(df.columns) if col in ['항목', '계']]

# '계'가 포함된 열 삭제
df_cleaned = df.drop(columns=df.columns[drop_columns])

# 첫 번째 행에 "계"가 포함된 열 삭제
df_cleaned = df_cleaned.loc[:, ~df_cleaned.iloc[0].str.contains('계')]


# 0행 삭제
df_cleaned = df_cleaned.drop(index=1)
# 결과 확인
df_cleaned
```

```python
# 첫 번째 열을 제외하고 나머지 열을 정수로 변환
a= df_cleaned.iloc[:, 1:] = df_cleaned.iloc[:, 1:].astype('int64')
a.info()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
```

```python
# 두 열씩 합치는 작업
columns = list(a.columns)
rows_count = a.shape[0]
combined_data = [[None for _ in range(10)] for _ in range(26)]

for i in range (rows_count):
  for j in range(0, len(columns)-1, 2):
      col1 = a.iloc[i][j]
      col2 = a.iloc[i][j + 1]
      k=j/2
      k=int(k)
      combined_data[i][k] = col1 + col2

df_sum = pd.DataFrame(combined_data)


# 새로운 열 이름을 위한 리스트 생성
new_column_names = [str(i) for i in range(2014,2024)]

# 열 이름 변경
df_sum.columns = new_column_names

# 자치구 추출
skipped_values = df_cleaned.iloc[:, 0].tolist()

# 선택된 값으로 df_sum 데이터프레임의 첫 번째 열 값 변경
df_sum.insert(loc=0,column='자치구',value= skipped_values)

#연도별 외국인 수 합계
df_all = df_sum.iloc[0]
df_all = pd.DataFrame(df_all)

# 행 인덱스를 첫 번째 열로 옮기기
df_all.reset_index(inplace=True)
df_all.rename(columns={df_all.columns[0]: '연도'}, inplace=True)
df_all.rename(columns={df_all.columns[1]: '합계'}, inplace=True)
df_all.drop(0, inplace=True)

# 첫 번째 열을 행 이름으로 설정
df_sum.set_index('자치구', inplace=True)

df_sum = df_sum.drop(index='합계')


# 결과 출력
print(df_sum)
df_all
```

```python
import pandas as pd

# 1. 행별 최댓값과 그 값의 열 이름
max_per_row = df_sum.idxmax(axis=1)  # 행별 최댓값의 열 이름
max_values = df_sum.max(axis=1)  # 행별 최댓값

min_per_row = df_sum.idxmin(axis=1)  # 행별 최솟값의 열 이름
min_values = df_sum.min(axis=1)  # 행별 최솟값

print("자치 구 별 최대 최소 5~9세 외국인 수:")
print(pd.DataFrame({'최댓값': max_values, '최댓값 연도': max_per_row,'최솟값': min_values, '최솟값 연도': min_per_row}))
df_gu = pd.DataFrame({'최댓값': max_values, '최댓값 연도': max_per_row,'최솟값': min_values, '최솟값 연도': min_per_row})

# 2. 열별 최댓값과 최솟값 그 값의 인덱스
max_per_column = df_sum.idxmax()  # 열별 최댓값의 인덱스
max_values_per_column = df_sum.max()  # 열별 최댓값

min_per_column = df_sum.idxmin()  # 열별 최솟값의 인덱스
min_values_per_column = df_sum.min()  # 열별 최솟값

print("\n2014~2023년도 별 5~9세 외국인 최대 최소 수와 자치 구 :")
print(pd.DataFrame({'최댓값': max_values_per_column, '최대 자치구': max_per_column, '최솟값': min_values_per_column, '최소 자치구': min_per_column}))
df_year = pd.DataFrame({'최댓값': max_values_per_column, '최대 자치구': max_per_column, '최솟값': min_values_per_column, '최소 자치구': min_per_column})
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 구성
districts = df_gu.index # '동별' 인덱스 제외
max_values = df_gu['최댓값']
min_values = df_gu['최솟값']

# 바 그래프 그리기
x = np.arange(len(districts))  # x 축 값 생성
width = 0.35  # 바 너비 설정

fig, ax = plt.subplots(figsize=(15,5))  # 그래프 크기 설정
bars1 = ax.bar(x - width/2, max_values, width, label='최댓값')  # 최댓값 바 그래프
bars2 = ax.bar(x + width/2, min_values, width, label='최솟값')  # 최솟값 바 그래프

# 라벨, 제목, 범례 설정
ax.set_xlabel('자치구 별')  # x축 라벨
ax.set_ylabel('인구 수')  # y축 라벨
ax.set_ylim(0,1600)
ax.set_title('2014~2023년도 서울 특별시 자치구별 최대 및 최소 5~14세 외국인 수')  # 그래프 제목
ax.set_xticks(x)  # x축 눈금 위치 설정
ax.set_xticklabels(districts, rotation=45)  # x축 라벨 회전
ax.legend()  # 범례 표시

plt.tight_layout()  # 레이아웃 조정
plt.show()  # 그래프 출력
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 구성
years = df_year.index  # 연도
max_values = df_year['최댓값']  # 최댓값
min_values = df_year['최솟값']  # 최솟값
max_districts = df_year['최대 자치구']  # 최대 구
min_districts = df_year['최소 자치구']  # 최소 구

# 바 그래프 그리기
x = np.arange(len(years))  # x 축 값 생성
width = 0.35  # 바 너비 설정


fig, ax = plt.subplots(figsize=(15,5))  # 그래프 크기 설정
bars1 = ax.bar(x - width/2, max_values, width, label='최댓값')  # 최댓값 바 그래프
bars2 = ax.bar(x + width/2, min_values, width, label='최솟값')  # 최솟값 바 그래프

# 구 이름 표시
for i in range(len(years)):
    ax.annotate(max_districts[i],
                xy=(x[i] - width/2, max_values[i]),
                xytext=(0, 3),  # 텍스트 위치 조정 (위로)
                textcoords="offset points",
                ha='center', va='bottom', fontsize=8)
    ax.annotate(min_districts[i],
                xy=(x[i] + width/2, min_values[i]),
                xytext=(0, 3),  # 텍스트 위치 조정 (위로)
                textcoords="offset points",
                ha='center', va='bottom', fontsize=8)

# 라벨, 제목, 범례 설정
ax.set_xlabel('연도')  # x축 라벨
ax.set_ylabel('인구 수')  # y축 라벨
ax.set_title('2014~2023년도 서울 특별시 최대 및 최소 외국인 수와 자치구')  # 그래프 제목
ax.set_xticks(x)  # x축 눈금 위치 설정
ax.set_xticklabels(years)  # x축 라벨
ax.legend()  # 범례 표시

plt.tight_layout()  # 레이아웃 조정
plt.show()  # 그래프 출력
```
